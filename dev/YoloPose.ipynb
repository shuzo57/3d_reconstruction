```python
from ultralytics import YOLO

import cv2
import matplotlib.pyplot as plt
```

```python
model = YOLO('../models/club.pt')
```

```python
img_path = '../data/img/red_0712_0016/red_0712_0016_1.jpg'
img = cv2.imread(img_path)
```

```python
fig, ax = plt.subplots()
ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax.axis('off')
plt.show()
```

```python
results = model(img, verbose=False)
```

```python
plot_img = results[0].plot()
```

```python
fig, ax = plt.subplots()
ax.imshow(cv2.cvtColor(plot_img, cv2.COLOR_BGR2RGB))
ax.axis('off')
plt.show()
```
